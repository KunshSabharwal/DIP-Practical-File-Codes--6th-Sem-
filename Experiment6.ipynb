```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters
from google.colab import files

uploaded = files.upload()

image_name = list(uploaded.keys())[0]

img = cv2.imread(image_name)

if img is None:
    raise ValueError("Image not loaded. Check filename or upload again.")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

laplacian = cv2.Laplacian(gray, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)

sobel = filters.sobel(gray)
prewitt = filters.prewitt(gray)
roberts = filters.roberts(gray)

sobel_h = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_v = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

plt.figure(figsize=(12,14))

plt.subplot(4,2,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(4,2,2)
plt.imshow(gray, cmap='gray')
plt.title("Gray Image")
plt.axis('off')

plt.subplot(4,2,3)
plt.imshow(laplacian, cmap='gray')
plt.title("Laplacian")
plt.axis('off')

plt.subplot(4,2,4)
plt.imshow(sobel, cmap='gray')
plt.title("Sobel")
plt.axis('off')

plt.subplot(4,2,5)
plt.imshow(prewitt, cmap='gray')
plt.title("Prewitt")
plt.axis('off')

plt.subplot(4,2,6)
plt.imshow(roberts, cmap='gray')
plt.title("Roberts")
plt.axis('off')

plt.subplot(4,2,7)
plt.imshow(np.abs(sobel_h), cmap='gray')
plt.title("Sobel Horizontal")
plt.axis('off')

plt.subplot(4,2,8)
plt.imshow(np.abs(sobel_v), cmap='gray')
plt.title("Sobel Vertical")
plt.axis('off')

plt.tight_layout()
plt.show()
```
