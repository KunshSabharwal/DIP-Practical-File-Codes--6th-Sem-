```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read two grayscale images
img1 = cv2.imread('test_image.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('test_image1.jpg', cv2.IMREAD_GRAYSCALE)

# Check if images are loaded
if img1 is None or img2 is None:
    print("Error: One or both images not found")
    exit()

# Ensure both images are of same size
if img1.shape != img2.shape:
    print("Error: Images must be of the same size")
    exit()

# Display both images
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title("Image 1")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img2, cmap='gray')
plt.title("Image 2")
plt.axis('off')

plt.show()

# Mean and Standard Deviation of Image 1
mean1 = np.mean(img1)
std1 = np.std(img1)

# Mean and Standard Deviation of Image 2
mean2 = np.mean(img2)
std2 = np.std(img2)

# Flatten images for correlation
img1_flat = img1.flatten()
img2_flat = img2.flatten()

# Correlation Coefficient
correlation = np.corrcoef(img1_flat, img2_flat)[0, 1]

# Display results
print("Image 1 Mean:", mean1)
print("Image 1 Standard Deviation:", std1)
print("Image 2 Mean:", mean2)
print("Image 2 Standard Deviation:", std2)
print("Correlation Coefficient:", correlation)
```
